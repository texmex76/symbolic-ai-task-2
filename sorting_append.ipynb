# Learning how to sort: training data

```python
import numpy as np
```

```python
sizes = [3, 4, 5]
```

Positive examples

```python
for size in sizes:
    for _ in range(400):
        listy = np.random.randint(0, 10, size)
        print(f"quick({list(listy)}, {sorted(listy)}).")
```

Negative examples

```python
for size in sizes:
    for _ in range(400):
        listy = np.random.randint(0, 10, size)
        if sorted(listy) != list(listy):
            print(f"quick({list(listy)}, {list(listy)}).")
```
